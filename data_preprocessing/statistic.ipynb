```python
import json
```

```python
data = json.load(open("data_target_ENSP.json"))
targets = {}
for d in data:
    target_list = d["target_ENSP"]
    for target in target_list:
        if target not in targets:
            targets[target] = 1
        targets[target] += 1
targets
```

```python
json.dump(targets, open("target_ENSP_count.json", "w"), indent=2)
```
