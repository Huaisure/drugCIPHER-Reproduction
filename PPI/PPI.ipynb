```python
# data preprocessing for PPI data
import pandas as pd

file_path = "9606.protein.links.v12.0.txt"
ppi_data = pd.read_csv(file_path, sep=' ', header=0)

# filter out the interactions with a combined score < threshold
threshold = 700
ppi_data = ppi_data[ppi_data['combined_score'] >= threshold]
ppi_data["protein1"] = ppi_data["protein1"].str.replace("9606.", "")
ppi_data["protein2"] = ppi_data["protein2"].str.replace("9606.", "")
ppi_data.to_csv("ppi_data_filtered.csv", index=False)
ppi_data
```

```python
# PPI network
import networkx as nx
import matplotlib.pyplot as plt

ppi_network = nx.Graph()
for _, row in ppi_data.iterrows():
    ppi_network.add_edge(row["protein1"], row["protein2"], weight=row["combined_score"])
ppi_network.number_of_nodes(), ppi_network.number_of_edges()
nx.write_graphml(ppi_network, "ppi_network.graphml")
```
